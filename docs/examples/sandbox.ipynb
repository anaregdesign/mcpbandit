```python
from mcpbandit.bandit import LinearContextArm, ThompsonSamplingPolicy
import numpy as np
```

```python
context_length = 5
arms = [LinearContextArm.of(context_length=context_length, lam=1) for _ in range(3)]

```

```python
policy = ThompsonSamplingPolicy(arms=arms, alpha=0.3)
```

```python
policy.select(context=np.array([1, 0, 0, 0, 0]))
```

```python
policy.observe(chosen_arm=0, reward=1.0, context=np.array([1, 0, 0, 0, 0]))

policy.observe(chosen_arm=1, reward=1.0, context=np.array([0, 1, 0, 0, 0]))
policy.observe(chosen_arm=2, reward=1.0, context=np.array([0, 0, 1, 0, 0]))
```

```python
(
    policy.select(context=np.array([1, 0, 0, 0, 0])),
    policy.select(context=np.array([0, 1, 0, 0, 0])),
    policy.select(context=np.array([0, 0, 1, 0, 0])),
    policy.select(context=np.array([0, 0, 0, 1, 0])),
)
```
